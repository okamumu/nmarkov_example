```python
import numpy as np
import scipy.sparse as sp
import scipy.io
import scipy.stats as stats
import scipy.linalg as lin
import matplotlib.pyplot as plt
import nmarkov as nm
```

```python
mat = scipy.io.loadmat('out.mat')
```

```python
Q=sp.csc_matrix(mat['Q'])
x=mat['x']
r=mat['r']
```

```python
plt.spy(Q, markersize=0.1)
```

```python
max(abs(Q.sum(axis=1)))
```

```python
ma = nm.nctmc(Q)
ma.params.rmax = 1000
```

```python
res=ma.trwd(t=np.linspace(0,100, num=10000), x0=x, rwd=r)
```

```python
plt.plot()
plt.plot(res['t'], res['irwd'])
```

```python
nm.sprob(Q) @ r
```
